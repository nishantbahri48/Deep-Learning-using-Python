```python
df = pd.read_excel("BankChurners.xlsx", index_col=0)
dataset = df.values
print(dataset)
X = df.customer_additions
Y = df.Avg_Utilization_Ratio
model = keras.models.Sequential()
model.add(keras.layers.Dense(200, activation='relu', input_shape=(1,)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(50, activation='relu'))
model.add(keras.layers.Dense(25, activation='relu'))
model.add(keras.layers.Dense(10, activation='relu'))
model.add(keras.layers.Dense(1))
model.compile(optimizer='adam', loss=tf.keras.losses.MeanAbsolutePercentageError())
```

```python
hist = model.fit(X, Y,batch_size=50, epochs=80000, validation_data=(X,Y))
```

```python
y_pred = model.predict(X)
```
